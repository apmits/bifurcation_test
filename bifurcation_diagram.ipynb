```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# define the logistic map    https://en.wikipedia.org/wiki/Logistic_map
def logistic(r, x):
    #        /\_ r is growth rate
    #            x is population
    return r * x * (1 - x)
```

```python
# define constants
R_ = 2
X_ = np.linspace(-50, 50, num=100)  # Return evenly spaced numbers over a specified interval. 
```

```python
X_
```

```python
# Draw the logistic function

fig, ax = plt.subplots(1, 1)

ax.plot(X_, logistic(R_, X_), color='k')  
#                                   'k' is black
```

```python
# Draw bifurcation diagram (for several r values)
```

```python
# using Python lists (seems faster than np.arrays with dtype='int64' !)

# r values to loop
r_range = np.linspace(-5, 5, num=100000)
# r_range = np.linspace(0, 4, int(1/0.0001))
len(r_range)
```

```python
r_range
```

```python
# Array of r values, for the x axis of the bifurcation plot
_r = []
# Array of x(t) values, for the y axis of the bifurcation plot
_x = []

for r in r_range:
    _r.append(r)
    # m = 0.1
    #   Start with a random value of m instead of remaining stuck
    #   on a particular branch of the diagram
    m = np.random.random()  # Return random floats in the half-open interval [0.0, 1.0).
    for n in range(1500):
      m = logistic(r, m)
      # m = (r*m)*(1-m)
    # Collection of data in Y must be done once per value of r
    _x.append(m)
    
```

```python
# Plot the data    
# plt.plot(_r, _x)  # won't do because this plot by default connects the datapoints with lines
plt.plot(_r, _x, ls='', marker=',')
plt.xlabel('r', fontsize=25)
plt.ylabel('x', fontsize=25)
# plt.figure(figsize=(1000, 500))
plt.rcParams['figure.figsize'] = [30, 20]
plt.show()
```
